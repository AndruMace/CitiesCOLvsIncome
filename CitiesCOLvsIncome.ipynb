```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='darkgrid', palette='deep')
```

```python
data = {'Cities': ['New York City', 'Los Angeles', 'Chicago', 'Houston', 'Pheonix', 'Philidelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
       'Average Income': [70295, 55909, 83890, 75377, 43960, 41449, 67980, 63400, 67382, 77000],
       'COL Index': [100.00, 78.20, 77.75, 63.62, 65.88, 80.19, 61.87, 72.67, 66.95, 76.36]}
```

```python
df = pd.DataFrame(data)
```

```python
cities_Plot = sns.scatterplot(x='Average Income', y='COL Index', data=df, hue='Cities', s=150)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
fig = cities_Plot.get_figure()
fig.savefig('CitiesColvIncome.png', bbox_inches='tight') 
```
